```python
import pandas as pd
import time

# ==========================================
# 1. LA LÓGICA DE NEGOCIO (EL PLANO)
# ==========================================
class ProspectoCliente():
    def __init__(self, nombre, presupuesto):
        self.nombre = nombre
        self.presupuesto = presupuesto
        self.intentos = 0
        self.estado = "FRIO 🥶"

    def procesar_cliente(self):
        # Simulamos que el robot hace su trabajo
        print(f"--- Procesando a: {self.nombre} ---")
        
        # Simulamos 3 llamadas rápidas
        for i in range(3):
            self.intentos += 1
            # REGLA DE NEGOCIO:
            # Si tiene más de $5,000 Y ya le llamamos 3 veces -> CALIENTE
            if self.presupuesto > 5000 and self.intentos >= 3:
                self.estado = "CALIENTE 🔥"
        
        print(f"   Resultado final: {self.estado}")
        
        # Devolvemos los datos finales para guardarlos después
        return {
            "Cliente": self.nombre,
            "Presupuesto": self.presupuesto,
            "Llamadas Hechas": self.intentos,
            "Estatus Final": self.estado
        }

# ==========================================
# 2. LA CONEXIÓN CON EL MUNDO REAL (EXCEL)
# ==========================================

print("📂 Leyendo archivo Excel 'cartera_clientes.xlsx'...")

try:
    # AQUI ESTÁ LA MAGIA: Pandas lee el Excel y lo convierte en una tabla
    df_entrada = pd.read_excel('cartera_clientes.xlsx')
    
    print(f"✅ Se encontraron {len(df_entrada)} clientes para procesar.\n")
    
    lista_resultados = []

    # 3. EL BUCLE DE TRABAJO
    # iterrows() nos permite ir fila por fila del Excel
    for index, fila in df_entrada.iterrows():
        
        # A. Creamos el Objeto con los datos de ESA fila del Excel
        # fila['Nombre'] busca la columna "Nombre" en el Excel
        robot_cliente = ProspectoCliente(fila['Nombre'], fila['Presupuesto'])
        
        # B. Ponemos al robot a trabajar
        datos_finales = robot_cliente.procesar_cliente()
        
        # C. Guardamos el reporte
        lista_resultados.append(datos_finales)
        time.sleep(0.5) # Pausa dramática

    # ==========================================
    # 4. EL REPORTE FINAL (OUTPUT)
    # ==========================================
    print("\n💾 Guardando resultados en un NUEVO Excel...")
    
    # Convertimos la lista de resultados en una tabla de Pandas
    df_salida = pd.DataFrame(lista_resultados)
    
    # Lo guardamos como un archivo nuevo (para no borrar el original)
    df_salida.to_excel("reporte_procesado.xlsx", index=False)
    
    print("✅ ¡Listo! Revisa el archivo 'reporte_procesado.xlsx'.")

except FileNotFoundError:
    print("❌ ERROR: No encontré el archivo 'cartera_clientes.xlsx'.")
    print("Asegúrate de crearlo y ponerlo en la misma carpeta que este código.")
```

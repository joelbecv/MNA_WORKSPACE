# REGION DE INTERES (REGION OF INTEREST, ROI)
### Una ROI permite elegir una sección de una imagen que es de interés y así reducir el esfuerzo del procesamiento computacional

#### https://pythonprogramming.net/lane-region-of-interest-python-plays-gta-v/

![Imagen](ROI_Example.png)

# Se cargan las bibliotecas necesarias

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Se muestra la imagen de prueba

```python
#se corrige el orden de los planos de color
#se obtiene la imagen en escala de grises y se muestra como arreglo
img_bgr = cv2.imread('image_lane_c.jpg')
img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_arr = np.asarray(img)
plt.imshow(img_arr, cmap='gray')
```

```python
#dimensiones de la image
img_arr.shape
```

# El primer paso es crear un polígono con la ROI y pintar de blanco la ROI dejando el resto de la imagen en negro

```python
# ROI arbitraria
vertices = np.array([[(20,540),(200, 300), (650, 300), (850,540)]], dtype=np.int32)    
# imagen en negro de las mismas dimensiones que la imagen original en escala de grises
roi_img = np.zeros_like(img)
# la ROI se pinta de blanco sobre la imagen en negro
cv2.fillPoly(roi_img, vertices, 255)
plt.imshow(roi_img, cmap='gray')
```

# El segundo paso es hacer una operación AND entre la imagen con la ROI y la imagen original en escala de grises

#### https://docs.opencv.org/4.x/d0/d86/tutorial_py_image_arithmetics.html
#### https://omes-va.com/operadores-bitwise/

![AND OPERATION](TABLA-DE-VERDAD-AND.png)

```python
img_con_roi = cv2.bitwise_and(img, roi_img)
plt.imshow(img_con_roi, cmap='gray')
```

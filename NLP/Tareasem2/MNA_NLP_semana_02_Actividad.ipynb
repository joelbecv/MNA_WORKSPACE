*   **Pregunta 1.** 

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario. 

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
# Pregunta 1: eliminar saltos de línea al final de cada comentario
docs_clean = [re.sub(r'\n$', '', doc) for doc in docs]
print(docs_clean[:10])
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!". 

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen. 

Indica cuántos resultados obtuviste.



```python
# Pregunta 2: palabras seguidas de 2 o más signos de admiración
resultado_p2 = re.findall(r'\w+!{2,}', ' '.join(docs_clean))
print(resultado_p2)
print(f'Total encontrados: {len(resultado_p2)}')
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
# Pregunta 3: palabras totalmente en mayúsculas (mínimo 2 letras)
resultado_p3 = re.findall(r'\b[A-Z]{2,}\b', ' '.join(docs_clean))
print(resultado_p3)
print(f'Total encontradas: {len(resultado_p3)}')
```

*   **Pregunta 4.**  

Busca e imprime los comentarios en donde todos los caracteres alfabéticos (letras) están en mayúsculas. 

Cada coincidencia encontrada debe ser todo el comentario/enunciado.

Indica cuántos resultados obtuviste.


```python
# Pregunta 4: comentarios donde todas las letras están en mayúsculas
resultado_p4 = [doc for doc in docs_clean if re.fullmatch(r'[^a-z]*', doc)]
print(resultado_p4)
print(f'Total encontrados: {len(resultado_p4)}')
```

*   **Pregunta 5.**  

Busca e imprime todas las palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú. 

Indica cuántos resultados obtuviste.

```python
# Pregunta 5: palabras con vocal acentuada (á é í ó ú)
resultado_p5 = re.findall(r'\w*[áéíóú]\w*', ' '.join(docs_clean), re.IGNORECASE)
print(resultado_p5)
print(f'Total encontradas: {len(resultado_p5)}')
```

*   **Pregunta 6.**  

Busca e imprime todas las cantidades numéricas monetarias, enteras o con decimales, que inician con el símbolo $\$$. 

Indica cuántos resultados obtuviste.

```python
# Pregunta 6: cantidades monetarias que inician con $
resultado_p6 = re.findall(r'\$\d+(?:\.\d+)?', ' '.join(docs_clean))
print(resultado_p6)
print(f'Total encontradas: {len(resultado_p6)}')
```

*   **Pregunta 7.**  

Busca e imprime todas las palabras que sean variantes de la palabra "love", sin importar si incluyen mayúsculas o minúsculas, o la manera en que esté conjugada o alguna otra variación que se haga con dicha palabra.

Indica cuántos resultados obtuviste.

```python
# Pregunta 7: variantes de la palabra 'love'
resultado_p7 = re.findall(r'\b\w*love\w*\b', ' '.join(docs_clean), re.IGNORECASE)
print(resultado_p7)
print(f'Total encontradas: {len(resultado_p7)}')
```

*   **Pregunta 8.**  

Busca e imprime todas las palabras, variantes de "so" y "good", que tengan dos o más "o" en "so" y 3 o más "o" en good. 

Indica cuántas encontraste.


```python
# Pregunta 8: variantes de 'so' con 2+ o's y 'good' con 3+ o's
texto = ' '.join(docs_clean)
variantes_so   = re.findall(r'\bso{2,}\b', texto, re.IGNORECASE)
variantes_good = re.findall(r'\bgo{3,}d\b', texto, re.IGNORECASE)
print('so:  ', variantes_so)
print('good:', variantes_good)
print(f'Total encontradas: {len(variantes_so) + len(variantes_good)}')
```

*   **Pregunta 9.**  

Busca e imprime todas las palabras que tengan una longitud mayor estrictamente a 10 caracteres alfabéticos.

No se consideran los signos de puntuación o caracteres especiales en la longitud de estas cadenas, solo caracteres alfabéticos en mayúsculas o minúsculas.

Indica la cantidad de palabras encontradas.


```python
# Pregunta 9: palabras con más de 10 caracteres alfabéticos
resultado_p9 = re.findall(r'\b[a-zA-Z]{11,}\b', ' '.join(docs_clean))
print(resultado_p9)
print(f'Total encontradas: {len(resultado_p9)}')
```

*   **Pregunta 10.**  

Busca e imprime todas las palabras que inician con una letra mayúscula y terminan con una minúscula, pero que además no sea la primera palabra del comentario/string. 

Indica la cantidad de resultados obtenidos.

```python
# Pregunta 10: palabras con inicial mayúscula y final minúscula que no sean la primera del comentario
resultado_p10 = []
for doc in docs_clean:
    palabras = re.findall(r'(?<=\s)[A-Z][a-zA-Z]*[a-z]\b', doc)
    resultado_p10.extend(palabras)
print(resultado_p10[:20])
print(f'Total encontradas: {len(resultado_p10)}')
```

*   **Pregunta 11.**  

Busca e imprime la secuencia de dos o más palabras que están separadas por un guion, "-", sin que tengan espacios en blanco entre ellas.

Por ejemplo "Go-Kart" sería válido, pero "Go  -Kart" o "Go  -  Kart" no lo serían. 

Indica la cantidad de resultados obtenidos.

```python
# Pregunta 11: secuencias de 2 o más palabras separadas por guion sin espacios
resultado_p11 = re.findall(r'\b\w+(?:-\w+)+\b', ' '.join(docs_clean))
print(resultado_p11)
print(f'Total encontradas: {len(resultado_p11)}')
```

*   **Pregunta 12.**  

Busca e imprime todas las palabras que terminan en "ing" o "ed". 

Indica la cantidad de palabras que encontraste de cada una.

```python
# Pregunta 12: palabras que terminan en 'ing' o 'ed'
texto = ' '.join(docs_clean)
palabras_ing = re.findall(r'\b\w+ing\b', texto, re.IGNORECASE)
palabras_ed  = re.findall(r'\b\w+ed\b',  texto, re.IGNORECASE)
print(f'Total "ing": {len(palabras_ing)}')
print(f'Total "ed":  {len(palabras_ed)}')
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes. 
   




```python
# Pregunta 13: limpieza — solo letras, minúsculas, sin espacios extra
docs_limpios = []
for doc in docs_clean:
    doc_limpio = re.sub(r'[^a-zA-Z\s]', ' ', doc)
    doc_limpio = doc_limpio.lower()
    doc_limpio = re.sub(r'\s+', ' ', doc_limpio).strip()
    docs_limpios.append(doc_limpio)
print(docs_limpios[:10])
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus. 

Es decir, al final de este proceso de tokenización, debes tener como resultado una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus. 

```python
# Pregunta 14: tokenización por palabras — lista de listas
tokens_corpus = [doc.split() for doc in docs_limpios]
total_tokens = sum(len(t) for t in tokens_corpus)
print(tokens_corpus[:3])
print(f'Total tokens en el corpus: {total_tokens}')
```

```python
# Pregunta 15: eliminar stopwords y calcular vocabulario
mis_stopwords_set = set(mis_stopwords)
corpus_sin_sw = [[tok for tok in doc if tok not in mis_stopwords_set] for doc in tokens_corpus]
total_sin_sw = sum(len(doc) for doc in corpus_sin_sw)
vocabulario  = set(tok for doc in corpus_sin_sw for tok in doc)
print(f'Tokens restantes después de eliminar stopwords: {total_sin_sw}')
print(f'Tokens únicos (vocabulario): {len(vocabulario)}')
```

Nivel de Significancia (alfa): Es un umbral predefinido antes de realizar la prueba de hipótesis, que representa la cantidad de casualidad o azar que estamos dispuestos a aceptar, Un valor común para alfa es 0.05 (5%), especialmente en entornos de negocio

**RESPUESTA:  PARA VARIABLES CATEGORICAS GENDER, EDUCATION**
Tampoco hay relación de estas variables con el la ausencia de datos en weight

**GENDER p-value of Chi_squared test: 0.5472167199163878**

**EDUCATION p-value of Chi_squared test: 0.8424822989541867**




```python
#ESTE BLOQUE VENÍA PRECARGADO Y NO SE CAMBIA NADA VS CÓDIGO DE LIBRO, ES EL PRIMER PASO PARA LOS CALCULOS DE RELACIÓN PARA VARIABLES NÚMERICAS
from scipy.stats import ttest_ind
def Diagnose_MV_Numerical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    labels=[]
    box_sr = pd.Series('',index = BM_MV.unique())
    for poss in BM_MV.unique():
        BM = BM_MV == poss
        box_sr[poss] = df[BM][str_att_name].dropna()
        labels.append(MV_labels[poss])

    plt.boxplot(box_sr,vert=False)
    plt.yticks([1,2],labels)
    plt.xlabel(str_att_name)
    plt.show()

    plt.figure(figsize=(8,3))

    att_range = (df[str_att_name].min(),df[str_att_name].max())

    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].hist()
        plt.xlim = att_range
        plt.xlabel(str_att_name)
        plt.title(MV_labels[poss])

    plt.show()

    group_1_data = df[BM_MV][str_att_name].dropna()
    group_2_data = df[~BM_MV][str_att_name].dropna()

    p_value = ttest_ind(group_1_data,group_2_data).pvalue

    print('p-value of t-test: {}'.format(p_value))
```

```python
#ESTE BLOQUE SE BASA TAMBIÉN EN LIBRO PERO SE MODIFICAN LOS ATRIBUTOS, EL DF Y VARIABLE DE REFERENCIA
numerical_attributes = ['Age', 'Number of siblings'] # ATRIBUTOS QUE CAMBIAN

BM_MV = response_df.Weight.isna() # ACÁ SE MODIFICA EL DF Y VARIABLE EN ESTE CASO PESO QUE ES LA QUE ESTAMOS ESTUDIANDO CON SUS MISSINGS
for att in numerical_attributes:
    print('Diagnosis Analysis of Missing Values for {}:'.format(att))
    Diagnose_MV_Numerical(response_df,att,BM_MV) #AQUÍ TAMBIÉN CAMBIA EL DF (SIN INDICAR VARIABLE)
    print('- - - - - - - - - - - - divider - - - - - - - - - - - ')
```

```python
#ESTE BLOQUE VENÍA PRECARGADO Y NO SE CAMBIA NADA VS CÓDIGO DE LIBRO, ES EL PRIMER PASO PARA LOS CALCULOS DE RELACIÓN PARA VARIABLES CATEGORICAS
from scipy.stats import chi2_contingency
def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    plt.figure(figsize=(10,4))
    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])
    p_value = chi2_contingency(contigency_table)[1]

    print('p-value of Chi_squared test: {}'.format(p_value))
```

```python
categorical_attributes = ['Gender', 'Education']

BM_MV = response_df.Weight.isna()
for att in categorical_attributes:
    print('Diagnosis Analysis of Missing Values for {}:'.format(att))
    Diagnose_MV_Categorical(response_df,att,BM_MV)
    print('- - - - - - - - - - - - divider - - - - - - - - - - - ')
```

7. ¿Hay valores atípicos de la variable `Weight`? Confírmalo con un gráfico de boxplot. ¿Cuántos son? Almacénalos en un dataframe independiente y elimínalos del conjunto original.

**RESPUESTA:SÍ HAY 10 OUTLIERS**

```python
plt.figure(figsize=(8, 3))
sns.boxplot(x=response_df['Weight'])
plt.show()
```

```python
# Calcula el IQR (rango intercuartílico)
Q1 = response_df['Weight'].quantile(0.25)
Q3 = response_df['Weight'].quantile(0.75)
IQR = Q3 - Q1

# Define los límites para los valores atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Identifica los valores atípicos
outliers = response_df[(response_df['Weight'] < lower_bound) | (response_df['Weight'] > upper_bound)]

# Almacenar los valores atípicos en un nuevo dataframe
outliers_df = outliers.copy()

```

```python
outliers_df
```

```python
# Contar cuántos valores atípicos hay
num_outliers = len(outliers_df)
print(f"Número de valores atípicos en Weight: {num_outliers}")
```

```python

# Eliminar los valores atípicos del dataframe original
response_df_cleaned = response_df[(response_df['Weight'] >= lower_bound) & (response_df['Weight'] <= upper_bound)].copy()

# Verificar las dimensiones del dataframe después de eliminar los atípicos
print(f"Dimensiones del dataframe original: {response_df.shape}")
print(f"Dimensiones del dataframe después de eliminar atípicos: {response_df_cleaned.shape}")
```

```python
plt.figure(figsize=(8, 3))
sns.boxplot(x=response_df_cleaned['Weight'])
plt.show()
```

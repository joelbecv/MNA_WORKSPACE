## **8-Ejecuta el siguiente código y comenta qué hace cada línea:**

1- datetime_df = air_df.ReadingDateTime.str.split(' ',expand=True) **R:Separar Fecha y Hora: Se toma la columna original que tiene la fecha y la hora juntas como texto, y se divide en dos partes: una con solo la fecha y otra con solo la hora. Estas dos partes se guardan aparte temporalmente.**

2- datetime_df.columns = ['Date','Time'] **R Nombrar Fecha y Hora: A las dos partes que se separaron en el paso anterior, se les ponen nombres: a la parte de la fecha se le llama 'Date' y a la parte de la hora se le llama 'Time'.**

3- datetime_df **R:Mostrar Fecha y Hora: Se muestra cómo quedaron esas dos partes ('Date' y 'Time') después de separarlas y nombrarlas.**

4- date_df = datetime_df.Date.str.split('/',expand=True)  **R: Separar Día, Mes y Año: Se toma la parte de la fecha ('Date') que se separó antes, y se vuelve a dividir usando las diagonales ('/') para obtener tres pedacitos: el día, el mes y el año. Estos tres pedacitos se guardan aparte temporalmente.**

5- date_df.columns = ['Day','Month','Year'] **R:Nombrar Día, Mes y Año: A los tres partes del paso anterior se les ponen nombres: 'Day' (Día), 'Month' (Mes) y 'Year' (Año).**

6- date_df **R: Mostrar Día, Mes y Año: se muestra cómo quedaron esas tres partes ('Day', 'Month', 'Year') después de separarlas y nombrarlas.**

7- air_df = air_df.join(date_df).join(datetime_df.Time).drop(columns=['ReadingDateTime','Year']) **R:Actualizar Tabla Principal: Se toman las partes útiles que se crearon ('Day', 'Month', 'Time') y se añaden como nuevas columnas a la tabla de datos principal. Al mismo tiempo, se eliminan de la tabla principal la columna original que tenía la fecha y hora juntas y también la columna 'Year' que se había extraído.**

8-air_df = air_df.set_index(['Month','Day','Time','Species']) **R: Reorganizar la Tabla (Índice): Se cambia la forma de organizar la tabla. En lugar de que cada fila tenga un número simple (0, 1, 2...), ahora cada fila se identifica usando una combinación de los valores de las columnas 'Month', 'Day', 'Time' y 'Species'. Es como darle una "dirección" más específica a cada fila usando esas cuatro características.**

9-air_df R: Mostrar Tabla Final:  **muestra cómo quedó la tabla principal de datos después de todos estos cambios: con las nuevas columnas de día, mes y hora, sin las columnas eliminadas, y organizada con la nueva forma de identificar las filas (el nuevo índice).**

```python
datetime_df = air_df.ReadingDateTime.str.split(' ',expand=True)
datetime_df.columns = ['Date','Time']
datetime_df
date_df = datetime_df.Date.str.split('/',expand=True)
date_df.columns = ['Day','Month','Year']
date_df
air_df = air_df.join(date_df).join(datetime_df.Time).drop(columns=['ReadingDateTime','Year'])
air_df = air_df.set_index(['Month','Day','Time','Species'])
air_df
```

## 10-Explica cuáles son las diferencias y similitudes entre el par .melt()/.pivot() y el par .stack()/.unstack(). Si tuvieras que elegir una contraparte para .melt() entre .stack()/.unstack(), ¿cuál elegirías? ¿Por qué?

Similitudes: Ambos pares de funciones sirven para cambiar la forma de un DataFrame, transformando entre formatos "largo" y "ancho".  .melt() y .stack() convierten un DataFrame ancho a largo,mientras que .pivot() y .unstack() hacen lo contrario, de largo a ancho.

Diferencias:

Especificidad: .melt() es más flexible para seleccionar columnas específicas para el "id_vars" (que permanecen como identificadores)# y "value_vars" (que se "derriten" en una sola columna).  .pivot() requiere especificar una columna para el índice, una para las columnas y una para los valores. .stack() y .unstack() operan sobre el índice jerárquico del DataFrame, es decir, los niveles del índice se mueven entre columnas y filas.  .stack() mueve los niveles del índice de columnas a filas, mientras que .unstack() hace lo contrario.


melt() vs .stack() / .unstack():

Si tuviera que elegir una contraparte para .melt() entre .stack() y .unstack(), elegiría .unstack().  

Razón:  Aunque no son perfectamente equivalentes, ambas funciones transforman un dataframe de un formato largo a uno ancho.  
.melt() toma columnas del DataFrame y las combina en dos columnas: una para identificar las observaciones y otra para los valores.
.unstack() toma niveles del índice jerárquico y los convierte en columnas.  En contextos donde se usa .melt() para convertir varias columnas a un formato largo, se necesitaría una operación análoga para revertirlo, para lo cual .unstack() es una buena opción.  
.stack() transformaría un DataFrame de formato ancho a largo, pero no directamente en la misma manera que .melt() lo hace, ya que trabajaría en columnas y no necesariamente en columnas específicas.
